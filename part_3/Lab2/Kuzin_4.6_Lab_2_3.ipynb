**Функции активации**

*Сигмоида*

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

def sigmoid_diff(x):
    return sigmoid(x)*(1-sigmoid(x))
```

*Th*

```python
def th(x):
    exp_x = np.exp(x)
    exp_dis_x= np.exp(-x)
    return (exp_x-exp_dis_x)/(exp_x+exp_dis_x)

def th_diff(x):
    return 1 - np.square(th(x))
```

*Softsign*

```python
def softsign(x):
    return x/(1+np.abs(x))

def softsign_diff(x):
    return 1/(1+np.abs(x))**2
```

*Relu*

```python
def relu(x):
    x[x<0] = 0
    return x

def relu_diff(x):
    x[x>=0] = 1
    x[x<0] = 0
    return x
```

```python
def prelu(x, a = 0.25):
    x[x<0] = a * x[x<0]
    return x

def prelu_diff(x, a = 0.25):
    x[x>=0] = 1
    x[x<0] = a
    return x
```

*ELU*

```python
def elu(x, a = 1):
    x[x<=0] = a * (np.exp(x[x<0])-1)
    return x

def elu_diff(x, a = 1):
    x[x>=0] = 1
    x[x<=0] = elu(x[x<=0], a) + a
    return x
```

*Linear function*

```python
def linear(weight_l, input_l, bias):
    return np.dot(weight_l, input_l) + bias
```

$$L = (Y_{1,m} - NN(X_{n,m}))^2.sum() = (Y_{1,m} - NN(X_{n,m}))^2\cdot E_{1,m},\ \ \  E_{1,m}=(1,1,\dots,1)^{\tau}_{1,m}$$

$$ \dfrac{\partial L}{\partial B_{k,1}}=  -2U_{1,k}^{\tau} ((\phi'(W_{k,n}\cdot X_{n,m} + B_{k,1}))\cdot (Y_{1,m} - NN(X_{n,m}))^{\tau})$$

*Градиентные спуски*

### SGD

$$W = W - \eta\dfrac{\partial L}{\partial W} $$

 $$\text{где } \dfrac{\partial L}{\partial W} \text{ рассчитывается отдельно для каждого батча}$$

### AdaGrad

$$G_i = G_{i-1} + \left(\dfrac{\partial L}{\partial W} \right)^2 $$

$$W = W - \dfrac{\eta}{\sqrt{G_i + \epsilon}}\dfrac{\partial L}{\partial W} $$

### Adam

$$M_i = \beta_1 M_{i-1} +(1-\beta_1)\dfrac{\partial L}{\partial W} $$

$$V_i = \beta_2 V_{i-1} +(1-\beta_2)\left(\dfrac{\partial L}{\partial W} \right)^2 $$

$$\hat{M}_i = \dfrac{M_i}{1-\beta_1^i} $$

$$\hat{V}_i = \dfrac{V_i}{1-\beta_2^i} $$

$$W = W - \dfrac{\eta}{\sqrt{\hat{V}_i + \epsilon}}\hat{M}_i $$


```python
from types import FunctionType

class NeuralNetwork:

    act_f_diff: FunctionType

    # Для Adagrad
    g_w1_l, g_w2_l, g_b1_l, g_b2_l = (0, 0, 0, 0)

    # Для Adam
    m_w1, m_w2, m_b1, m_b2 = (0, 0, 0, 0)
    v_w1, v_w2, v_b1, v_b2 = (0, 0, 0, 0)

    def __init__(self, x, y, n = 5, act_f = sigmoid): # x - исходные значения (датасет), y - target значения
        self.input = x # входные данные
        self.target = y # целевые(выходные) значения
        self.n_hidden_neurons = n # кол-во нейронов в слое
        self.act_f = act_f # функция активации

        # Инициализируем производную функции активации
        if act_f == sigmoid:
            self.act_f_diff = sigmoid_diff
        elif act_f == th:
            self.act_f_diff = th_diff
        elif act_f == softsign:
            self.act_f_diff = softsign_diff
        elif act_f == relu:
            self.act_f_diff = relu_diff
        elif act_f == elu:
            self.act_f_diff = elu_diff
        else:
            raise("Это не функция активации!")

        # Функции линейных слоев
        self.fc1 = linear
        self.bias1 = np.random.rand(self.n_hidden_neurons, 1)*0.99+0.01
        self.fc2 = linear
        self.bias2 = np.random.rand()*0.99+0.01

        # Инициализируем веса и выходные значения рандомными числами
        self.weights1 = np.random.rand(self.n_hidden_neurons, self.input.shape[0])*0.99+0.01
        self.weights2 = np.random.rand(1, self.n_hidden_neurons)*0.99+0.01
        self.output = np.zeros(self.target.shape)

        # Обновляем сумму градиентов
        self.g_w1_l = np.zeros(self.weights1.shape)
        self.g_w2_l = np.zeros(self.weights2.shape)
        self.g_b1_l = np.zeros(self.bias1.shape)
        self.g_b2_l = 0.

        # self.m_w1, self.m_w2, self.m_b1, self.m_b2 = np.zeros((4, n, 1))
        # self.v_w1, self.v_w2, self.v_b1, self.v_b2 = np.zeros((4, n, 1))

    def forward(self, x, y = None):
        self.input = x
        if y is not None:
            self.target = y
        self.layer1 = self.fc1(self.weights1, self.input, self.bias1)
        self.act_f1 = self.act_f(self.layer1)
        self.output = self.fc2(self.weights2, self.act_f1, self.bias2)
        return self.output

    def backprop(self, epoch=1, grad = 'SGD', alpha = None):
        if alpha == None:
            if grad == 'Adam':
                self.adam(self.input, epoch=epoch)
            elif grad == 'SGD':
                self.sgd(self.input)
            elif grad == 'Adagrad':
                self.adagrad(self.input)
        else:
            if grad == 'Adam':
                self.adam(self.input, epoch=epoch, a = alpha)
            elif grad == 'SGD':
                self.sgd(self.input, a = alpha)
            elif grad == 'Adagrad':
                self.adagrad(self.input, a = alpha)

    # Градиентные шаги
    def sgd(self, x, a = 1.0e-2):

        d_weights2 = 2.*np.dot((self.target - self.output), self.act_f1.T)
        d_weights1 = 2.*self.weights2.T*np.dot((self.target - self.output)*self.act_f_diff(self.layer1), x.T)
        d_bias1    = 2.*self.weights2.T*np.dot(self.act_f_diff(self.layer1), (self.target - self.output).T)
        d_bias2    = 2.*(self.target - self.output).sum()

        self.weights1 += a * d_weights1/self.input.shape[1]
        self.weights2 += a * d_weights2/self.input.shape[1]
        self.bias1    += a * d_bias1/self.input.shape[1]
        self.bias2    += a * d_bias2 /self.input.shape[1]

    # Адаптивный градиентный спуск
    def adagrad(self, x, a = 1.0e-2, eps = 1.0e-9):
        # Градиенты
        d_weights2 = 2*np.dot((self.target - self.output), self.act_f1.T)
        d_weights1 = 2*self.weights2.T*np.dot((self.target - self.output)*self.act_f_diff(self.layer1), x.T)
        d_bias1    = 2*self.weights2.T*np.dot(self.act_f_diff(self.layer1), (self.target - self.output).T)
        d_bias2    = 2*(self.target - self.output).sum()

        # lr_s = lr/(1+(epoch-1))

        # Обновляем сумму градиентов
        self.g_w1_l += np.square(d_weights1)
        self.g_w2_l += np.square(d_weights2)
        self.g_b1_l += np.square(d_bias1)
        self.g_b2_l += np.square(d_bias2)

        self.weights1 += a * d_weights1/ np.sqrt(self.g_w1_l + eps)/self.input.shape[1]
        self.weights2 += a * d_weights2 / np.sqrt(self.g_w2_l + eps)/self.input.shape[1]
        self.bias1 += a * d_bias1 / (np.sqrt(self.g_b1_l) + eps)/self.input.shape[1]
        self.bias2 += a * d_bias2 / (np.sqrt(self.g_b2_l) + eps)/self.input.shape[1]

    # Adam
    def adam(self, x, epoch, a = 1.0e-3, beta1 = 0.9, beta2 = 0.9, eps = 1.0e-9):

        # Градиенты
        d_weights2 = 2*np.dot((self.target - self.output), self.act_f1.T)
        d_weights1 = 2*self.weights2.T*np.dot((self.target - self.output)*self.act_f_diff(self.layer1), x.T)
        d_bias1    = 2*self.weights2.T*np.dot(self.act_f_diff(self.layer1), (self.target - self.output).T)
        d_bias2    = 2*(self.target - self.output).sum()

        # Обновляем основу первого и второго моментов импульсов
        self.m_w1, self.m_w2, self.m_b1, self.m_b2 = list(map(lambda grad, m: (1-beta1)*grad+beta1*m ,(d_weights1, d_weights2, d_bias1, d_bias2), (self.m_w1, self.m_w2, self.m_b1, self.m_b2)))
        m_corrected_w1, m_corrected_w2, m_corrected_b1, m_corrected_b2  = list(map(lambda m: m / (1-np.power(beta1, epoch)),(self.m_w1, self.m_w2, self.m_b1, self.m_b2)))

        self.v_w1, self.v_w2, self.v_b1, self.v_b2 = list(map(lambda grad, v: (1-beta2)*grad**2+beta2*v,(d_weights1, d_weights2, d_bias1, d_bias2), (self.v_w1, self.v_w2, self.v_b1, self.v_b2)))
        v_corrected_w1, v_corrected_w2, v_corrected_b1, v_corrected_b2 = list(map(lambda v: v/ (1-np.power(beta2, epoch)),(self.v_w1, self.v_w2, self.v_b1, self.v_b2)))

        self.weights1 += a*m_corrected_w1 / (np.sqrt(v_corrected_w1) + eps)/self.input.shape[1]
        self.weights2 += a*m_corrected_w2 / (np.sqrt(v_corrected_w2) + eps)/self.input.shape[1]
        self.bias1 += a*m_corrected_b1 / (np.sqrt(v_corrected_b1) + eps)/self.input.shape[1]
        self.bias2 += a*m_corrected_b2 / (np.sqrt(v_corrected_b2) + eps)/self.input.shape[1]

    def test(self, t):
        return self.fc2(self.weights2, self.act_f(self.fc1(self.weights1, t, self.bias1)), self.bias2)
```

```python
def loss(pred, target):
    squares = ((pred - target) ** 2)**(0.5)
    return squares.mean()
```

  \right)
   $$

$$NN(X) = U\psi (V\cdot\phi(W\cdot X+B) + C) + d=\left(
  \begin{array}{ccc}
  \dots & \dots & \dots \\
  \dots & u_r\psi\left(\sum\limits_{i=1}^{k}v_{ri}\phi\left(\sum\limits_{l=1}^{n}w_{il}x_{lp} + b_{i}\right) +c_r\right) + d  & \dots \\
  \dots & \dots & \dots \\
  \end{array}
  \right)
   $$


$$L(X_p) = \sum\limits_{p=1}^{m}(y_p-NN(X_p))^2 $$

$$ \dfrac{\partial L}{\partial u_{r}}= -2 \sum\limits_{p=1}^{m}(y_p-NN(X_p))  \psi\left(\sum\limits_{i=1}^{k}v_{ri}\phi\left(\sum\limits_{l=1}^{n}w_{il}x_{lp} + b_{i}\right) +c_r\right) $$


$$ \dfrac{\partial L}{\partial v_{ri}}= -2 \sum\limits_{p=1}^{m}(y_p-NN(X_p))  u_r\psi'\left(\sum\limits_{i=1}^{k}v_{ri}\phi\left(\sum\limits_{l=1}^{n}w_{il}x_{lp} + b_{i}\right) +c_r\right)\phi\left(\sum\limits_{l=1}^{n}w_{il}x_{lp} + b_{i}\right) $$


$$ \dfrac{\partial L}{\partial w_{il}}= -2 \sum\limits_{p=1}^{m}(y_p-NN(X_p))  u_r\psi'\left(\sum\limits_{i=1}^{k}v_{ri}\phi\left(\sum\limits_{l=1}^{n}w_{il}x_{lp} + b_{i}\right) +c_r\right)v_{ri}\phi'\left(\sum\limits_{l=1}^{n}w_{il}x_{lp} + b_{i}\right)x_{lp}$$

$$ \dfrac{\partial L}{\partial c_{r}}= -2 \sum\limits_{p=1}^{m}(y_p-NN(X_p))  u_r\psi'\left(\sum\limits_{i=1}^{k}v_{ri}\phi\left(\sum\limits_{l=1}^{n}w_{il}x_{lp} + b_{i}\right) +c_r\right) $$

$$ \dfrac{\partial L}{\partial b_{i}}= -2 \sum\limits_{p=1}^{m}(y_p-NN(X_p))  u_r\psi'\left(\sum\limits_{i=1}^{k}v_{ri}\phi\left(\sum\limits_{l=1}^{n}w_{il}x_{lp} + b_{i}\right) +c_r\right)v_{ri}\phi'\left(\sum\limits_{l=1}^{n}w_{il}x_{lp} + b_{i}\right)$$

$$ \dfrac{\partial L}{\partial d}= -2 \sum\limits_{p=1}^{m}(y_p-NN(X_p))$$

$$ \dfrac{\partial L}{\partial U_{1,s}}= -2 (Y_{1,m} - NN(X_{n,m})) \cdot  \psi (V_{s,k}\cdot\phi(W_{k,n}\cdot X_{n,m} + B_{k,1}) + C_{s,1}) ^{\tau}$$

$$ \dfrac{\partial L}{\partial V_{s,k}}= -2U_{1,s} ^{\tau}(( (Y_{1,m} - NN(X_{n,m}))   \psi' (V_{s,k}\cdot\phi(W_{k,n}\cdot X_{n,m} + B_{k,1}) + C_{s,1}))\cdot\phi(W_{k,n}\cdot X_{n,m} + B_{k,1}) ^{\tau})$$


$$ \dfrac{\partial L}{\partial W_{k,n}}=  -2(((U_{1,s}V_{s,k}^{\tau}) \cdot ( (Y_{1,m} - NN(X_{n,m}))   \psi' (V_{s,k}\cdot\phi(W_{k,n}\cdot X_{n,m} + B_{k,1}) + C_{s,1}) ))\phi'(W_{k,n}\cdot X_{n,m} + B_{k,1})))\cdot X_{n,m} ^{\tau}$$

$$ \dfrac{\partial L}{\partial B_{k,1}}=   -2(((U_{1,s}V_{s,k}^{\tau}) \cdot (    \psi' (V_{s,k}\cdot\phi(W_{k,n}\cdot X_{n,m} + B_{k,1}) + C_{s,1}) ))\phi'(W_{k,n}\cdot X_{n,m} + B_{k,1})))\cdot (Y_{1,m} - NN(X_{n,m})) ^{\tau}$$

$$ \dfrac{\partial L}{\partial C_{s,1}}=  -2U_{1,s}^{\tau}  (\psi' (V_{s,k}\cdot\phi(W_{k,n}\cdot X_{n,m} + B_{k,1}) + C_{s,1})\cdot ((Y_{1,m} - NN(X_{n,m}))^{\tau} )  $$



```python
from types import FunctionType

class NeuralNetwork:

    act_f_diff: FunctionType

    # Для Adagrad
    g_w1_l, g_w2_l, g_w3_l, g_b1_l, g_b2_l, g_b3_l = (0, 0, 0, 0, 0, 0)

    # Для Adam
    m_w1, m_w2, m_w3, m_b1, m_b2, m_b3 = (0, 0, 0, 0, 0, 0)
    v_w1, v_w2, v_w3, v_b1, v_b2, v_b3 = (0, 0, 0, 0, 0, 0)


    def __init__(self, x, y, n1 = 5, n2 = 5, act_f = th, act_f2 = th): # x - исходные значения (датасет), y - target значения
        self.input = x # входные данные
        self.target = y # целевые(выходные) значения
        self.n_hidden_neurons_1 = n1 # кол-во нейронов в 1 слое
        self.n_hidden_neurons_2 = n2 # кол-во нейронов в 2 слое
        self.act_f = act_f # функция активации
        self.act_f2 = act_f2

        # Инициализируем производную функции активации
        if act_f == sigmoid:
            self.act_f_diff = sigmoid_diff
        elif act_f == th:
            self.act_f_diff = th_diff
        elif act_f == softsign:
            self.act_f_diff = softsign_diff
        elif act_f == relu:
            self.act_f_diff = relu_diff
        elif act_f == prelu:
            self.act_f_diff = prelu_diff
        elif act_f == elu:
            self.act_f_diff = elu_diff
        else:
            raise("Это не функция активации!")

        # Инициализируем производную функции активации
        if act_f2 == sigmoid:
            self.act_f_diff2 = sigmoid_diff
        elif act_f2 == th:
            self.act_f_diff2 = th_diff
        elif act_f2 == softsign:
            self.act_f_diff2 = softsign_diff
        elif act_f2 == relu:
            self.act_f_diff2 = relu_diff
        elif act_f2 == prelu:
            self.act_f_diff2 = prelu_diff
        elif act_f2 == elu:
            self.act_f_diff2 = elu_diff
        else:
            raise("Это не функция активации!")

        # Функции линейных слоев
        self.fc1 = linear
        self.bias1 = np.random.rand(self.n_hidden_neurons_1, 1)*0.99+0.01
        self.fc2 = linear
        self.bias2 = np.random.rand(self.n_hidden_neurons_2, 1)*0.99+0.01
        self.fc3 = linear
        self.bias3 = np.random.rand()*0.99+0.01

        # Инициализируем веса и выходные значения рандомными числами
        self.weights1 = np.random.rand(self.n_hidden_neurons_1, self.input.shape[0])*0.99+0.01
        self.weights2 = np.random.rand(self.n_hidden_neurons_2, self.n_hidden_neurons_1)*0.99+0.01
        self.weights3 = np.random.rand(1, self.n_hidden_neurons_2)*0.99+0.01
        self.output = np.zeros(self.target.shape)

        # # Обновляем сумму градиентов
        # self.g_w1_l = np.zeros(self.weights1.shape)
        # self.g_w2_l = np.zeros(self.weights2.shape)
        # self.g_b1_l = np.zeros(self.bias1.shape)
        # self.g_b2_l = 0

        # self.m_w1, self.m_w2, self.m_b1, self.m_b2 = np.zeros((4, n, 1))
        # self.v_w1, self.v_w2, self.v_b1, self.v_b2 = np.zeros((4, n, 1))

    def forward(self, x, y = None):
        self.input = x
        if y is not None:
            self.target = y
        self.layer1 = self.fc1(self.weights1, self.input, self.bias1)
        self.act_f1 = self.act_f(self.layer1)
        self.layer2 = self.fc2(self.weights2, self.act_f1, self.bias2)
        self.act_f_2 = self.act_f2(self.layer2)
        self.output = self.fc3(self.weights3, self.act_f_2, self.bias3)
        return self.output

    def backprop(self, epoch=1, grad = 'SGD', alpha = None):
        if alpha == None:
            if grad == 'Adam':
                self.adam(self.input, epoch=epoch)
            elif grad == 'SGD':
                self.sgd(self.input)
            elif grad == 'Adagrad':
                self.adagrad(self.input)
        else:
            if grad == 'Adam':
                self.adam(self.input, epoch=epoch, a = alpha)
            elif grad == 'SGD':
                self.sgd(self.input, a = alpha)
            elif grad == 'Adagrad':
                self.adagrad(self.input, a = alpha)

    # Градиентные шаги
    def sgd(self, x, a = 0.001):

        d_weights3 = 2*np.dot((self.target - self.output), self.act_f_2.T)
        d_weights2 = 2*self.weights3.T * np.dot((self.target - self.output) * self.act_f_diff2(self.layer2), self.act_f1.T)
        d_weights1 = 2*np.dot(np.dot(self.weights3*self.weights2.T, (self.target - self.output) * self.act_f_diff2(self.layer2))*self.act_f_diff(self.layer1), self.input.T)

        d_bias3    = 2*(self.target - self.output).sum()
        d_bias2    = 2*self.weights3.T*np.dot(self.act_f_diff2(self.layer2), (self.target - self.output).T)
        d_bias1    = 2*np.dot(np.dot(self.weights3*self.weights2.T,self.act_f_diff2(self.layer2))*self.act_f_diff2(self.layer1), (self.target - self.output).T)

        self.weights1 += a * d_weights1/self.input.shape[1]
        self.weights2 += a * d_weights2/self.input.shape[1]
        self.weights3 += a * d_weights3/self.input.shape[1]
        self.bias1    += a * d_bias1/self.input.shape[1]
        self.bias2    += a * d_bias2/self.input.shape[1]
        self.bias3    += a * d_bias3/self.input.shape[1]

    # Адаптивный градиентный спуск
    def adagrad(self, x, a = 1.0e-1, eps = 1.0e-9):

        # Градиенты
        d_weights3 = 2*np.dot((self.target - self.output), self.act_f_2.T)
        d_weights2 = 2*self.weights3.T * np.dot((self.target - self.output) * self.act_f_diff(self.layer2), self.act_f1.T)
        d_weights1 = 2*np.dot(np.dot(self.weights3*self.weights2.T, (self.target - self.output) * self.act_f_diff(self.layer2))*self.act_f_diff(self.layer1), self.input.T)

        d_bias3    = 2*(self.target - self.output).sum()
        d_bias2    = 2*self.weights3.T*np.dot(self.act_f_diff(self.layer2), (self.target - self.output).T)
        d_bias1    = 2*np.dot(np.dot(self.weights3*self.weights2.T,self.act_f_diff(self.layer2))*self.act_f_diff(self.layer1), (self.target - self.output).T)
        #  2*np.dot(np.dot(self.weights3*self.weights2.T,sigmoid_diff(self.layer2))*sigmoid_diff(self.layer1), (self.y - self.output).T)

        # lr_s = lr/(1+(epoch-1))

        # Обновляем сумму градиентов
        self.g_w1_l += np.square(d_weights1)
        self.g_w2_l += np.square(d_weights2)
        self.g_w3_l += np.square(d_weights3)
        self.g_b1_l += np.square(d_bias1)
        self.g_b2_l += np.square(d_bias2)
        self.g_b3_l += np.square(d_bias3)

        self.weights1 += a * d_weights1 / np.sqrt(self.g_w1_l + eps)/self.input.shape[1]
        self.weights2 += a * d_weights2 / np.sqrt(self.g_w2_l + eps)/self.input.shape[1]
        self.weights3 += a * d_weights3 / np.sqrt(self.g_w3_l + eps)/self.input.shape[1]
        self.bias1 += a * d_bias1 / (np.sqrt(self.g_b1_l) + eps)/self.input.shape[1]
        self.bias2 += a * d_bias2 / (np.sqrt(self.g_b2_l) + eps)/self.input.shape[1]
        self.bias3 += a * d_bias3 / (np.sqrt(self.g_b3_l) + eps)/self.input.shape[1]

    # Adam
    def adam(self, x, epoch, a = 1.0e-2, beta1 = 0.9, beta2 = 0.9, eps = 1.0e-9):

        # Градиенты
        d_weights3 = 2*np.dot((self.target - self.output), self.act_f_2.T)
        d_weights2 = 2*self.weights3.T * np.dot((self.target - self.output) * self.act_f_diff2(self.layer2), self.act_f1.T)
        d_weights1 = 2*np.dot(np.dot(self.weights3*self.weights2.T, (self.target - self.output) * self.act_f_diff2(self.layer2))*self.act_f_diff(self.layer1), self.input.T)

        d_bias3    = 2*(self.target - self.output).sum()
        d_bias2    = 2*self.weights3.T*np.dot(self.act_f_diff2(self.layer2), (self.target - self.output).T)
        d_bias1    = 2*np.dot(np.dot(self.weights3*self.weights2.T,self.act_f_diff2(self.layer2))*self.act_f_diff2(self.layer1), (self.target - self.output).T)

        # Обновляем основу первого и второго моментов импульсов
        self.m_w1, self.m_w2, self.m_w3, self.m_b1, self.m_b2, self.m_b3 = list(map(lambda grad, m: (1-beta1)*grad+beta1*m ,(d_weights1, d_weights2, d_weights3, d_bias1, d_bias2, d_bias3), (self.m_w1, self.m_w2, self.m_w3, self.m_b1, self.m_b2, self.m_b3)))
        m_corrected_w1, m_corrected_w2, m_corrected_w3, m_corrected_b1, m_corrected_b2, m_corrected_b3  = list(map(lambda m: m / (1-np.power(beta1, epoch)),(self.m_w1, self.m_w2, self.m_w3, self.m_b1, self.m_b2, self.m_b3)))

        self.v_w1, self.v_w2, self.v_w3, self.v_b1, self.v_b2, self.v_b3 = list(map(lambda grad, v: (1-beta2)*grad**2+beta2*v,(d_weights1, d_weights2, d_weights3, d_bias1, d_bias2, d_bias3), (self.v_w1, self.v_w2, self.v_w3, self.v_b1, self.v_b2, self.v_b3)))
        v_corrected_w1, v_corrected_w2, v_corrected_w3, v_corrected_b1, v_corrected_b2, v_corrected_b3 = list(map(lambda v: v/ (1-np.power(beta2, epoch)),(self.v_w1, self.v_w2, self.v_w3, self.v_b1, self.v_b2, self.v_b3)))

        self.weights1 += a*m_corrected_w1 / (np.sqrt(v_corrected_w1) + eps)/self.input.shape[1]
        self.weights2 += a*m_corrected_w2 / (np.sqrt(v_corrected_w2) + eps)/self.input.shape[1]
        self.weights3 += a*m_corrected_w3 / (np.sqrt(v_corrected_w3) + eps)/self.input.shape[1]
        self.bias1 += a*m_corrected_b1 / (np.sqrt(v_corrected_b1) + eps)/self.input.shape[1]
        self.bias2 += a*m_corrected_b2 / (np.sqrt(v_corrected_b2) + eps)/self.input.shape[1]
        self.bias3 += a*m_corrected_b3 / (np.sqrt(v_corrected_b3) + eps)/self.input.shape[1]
```

*Loss-функция*

```python
def loss(pred, target):
    squares = ((pred - target) ** 2)**(0.5)
    return squares.mean()
```
