```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

from mpl_toolkits.mplot3d.axes3d import get_test_data

```

```python
def f(x,y):
    return (np.sin(y/6)/(1 + x**2/10) + 1)/2
```

$f(x_0+a_1t,y_0+a_2t) $ 

$\overrightarrow{a}=(a_1,a_2), \ \sqrt{a_1^2+a_2^2}=|\overrightarrow{a}|=1 $

$f'_t= \dfrac{\partial f}{\partial x}\dfrac{\partial x}{\partial t}+ \dfrac{\partial f}{\partial y}\dfrac{\partial y}{\partial t}=$


$= \dfrac{\partial f}{\partial x}a_1+\dfrac{\partial f}{\partial y}a_2=\langle \nabla f, \overrightarrow{a} \rangle=|\nabla f|\cdot|\overrightarrow{a}|\cos(\nabla f, \overrightarrow{a})=|\nabla f|\cdot\cos(\nabla f, \overrightarrow{a})$

$\overrightarrow{a}=-\nabla f$

```python
def f_grad(x,y):
    return  -(np.sin(y/6)*x/5)/(1 + x**2/10)**2/2, (np.cos(y/6)/(1 + x**2/10))/12
```

$\left(-\dfrac{\sin(y/6)x/5}{(1+x^2/10)}\cdot 0,5, \dfrac{\cos(y/6)/6}{(1+x^2/10)}\cdot 0,5 \right)$

```python
fig = plt.figure(figsize=plt.figaspect(0.5), dpi=200)

ax = fig.add_subplot(1, 2, 1, projection='3d')
X = np.arange(-20, 20, 0.25)
Y = np.arange(-20, 20, 0.25)
X, Y = np.meshgrid(X, Y)
Z = f(X,Y)
x_0 = 12
y_0 = 3
for i in np.arange(20000):
    z_0 = f(x_0,y_0)
    x_1 = x_0 - f_grad(x_0,y_0)[0]
    y_1 = y_0 - f_grad(x_0,y_0)[1]
    z_1 = f(x_1,y_1)
    ax.plot([x_0, x_1], [y_0, y_1], [z_0, z_1], color='black',zorder=1)
    x_0, y_0 = x_1, y_1
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm, linewidth=0, antialiased=False)
ax = fig.add_subplot(1, 2, 2, projection='3d')

ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)
```

$f_2(x,y) = \left(\dfrac{\sin(y/6)}{(1+x^2/10+0,01y^2)} + 1\right)\cdot 0,5$

```python
def f_2(x,y):
    return (np.sin(y/6)/(1 + x**2/10+0.001*y**2) + 1)/2
```

```python
def f_2_grad(x,y):
    return  -((np.sin(y/6)*x)/(1 + x**2/10+0.001*y**2)**2)/10, ((np.cos(y/6)*(1 + x**2/10+0.001*y**2)-np.sin(y/6)*12*y*0.001)/(1 + x**2/10+0.001*y**2)**2)/12
```

```python
fig = plt.figure(figsize=plt.figaspect(0.5), dpi=200)

ax = fig.add_subplot(1, 2, 1, projection='3d')
#ax.scatter(x_0, y_0, z_0, color='red',zorder=10)
X = np.arange(-60, 60, 0.25)
Y = np.arange(-60, 60, 0.25)
X, Y = np.meshgrid(X, Y)
Z = f_2(X,Y)
x_0 = 7
y_0 = -35
for i in np.arange(1000):
    z_0 = f_2(x_0,y_0)
    x_1 = x_0 - 10*f_2_grad(x_0,y_0)[0]
    y_1 = y_0 - 10*f_2_grad(x_0,y_0)[1]
    z_1 = f_2(x_1,y_1)
    ax.plot([x_0, x_1], [y_0, y_1], [z_0, z_1], color='black',zorder=10)
    x_0, y_0 = x_1, y_1
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm, linewidth=0, antialiased=False)
ax = fig.add_subplot(1, 2, 2, projection='3d')

ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)
```

$$ g(x,y)=x^2+0,01y^2$$

```python
def g(x,y):
    return x**2+0.01*y**2
```

```python
def g_grad(x,y):
    return 2*x, 0.02*y
```

```python
fig = plt.figure(figsize=plt.figaspect(0.5), dpi=200)

ax = fig.add_subplot(1, 2, 1, projection='3d')
#ax.scatter(x_0, y_0, z_0, color='red',zorder=10)
X = np.arange(-2, 2, 0.25)
Y = np.arange(-2, 2, 0.25)
X, Y = np.meshgrid(X, Y)
Z = g(X,Y)
x_0 = -2
y_0 = -1.5
for i in np.arange(1000):
    z_0 = g(x_0,y_0)
    x_1 = x_0 - g_grad(x_0,y_0)[0]
    y_1 = y_0 - g_grad(x_0,y_0)[1]
    z_1 = g(x_1,y_1)
    ax.plot([x_0, x_1], [y_0, y_1], [z_0, z_1], color='black',zorder=10)
    x_0, y_0 = x_1, y_1
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm, linewidth=0, antialiased=False)
ax = fig.add_subplot(1, 2, 2, projection='3d')

ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)
```

$\overrightarrow{a}=(a_1,a_2,...), \  \overrightarrow{b}=(b_1,b_2,...),$

$$\cos (\overrightarrow{a},\overrightarrow{b}) =\dfrac{\sum\limits_{i=1}^na_ib_i} {\sqrt{\sum\limits_{i=1}^na_i^2} \sqrt{\sum\limits_{i=1}^nb_i^2}}  $$

```python
def Cos(a,b):
    return np.dot(a,b)/(np.linalg.norm(a)*np.linalg.norm(b))
```

```python
fig = plt.figure(figsize=plt.figaspect(0.5), dpi=200)
alpha = 0.99
ax = fig.add_subplot(1, 2, 1, projection='3d')
#ax.scatter(x_0, y_0, z_0, color='red',zorder=10)
X = np.arange(-2, 2, 0.25)
Y = np.arange(-2, 2, 0.25)
X, Y = np.meshgrid(X, Y)
Z = g(X,Y)
x_0 = -2
y_0 = -1.5
x_1 = 0
y_1 = 0
for i in np.arange(10):
    z_0 = g(x_0,y_0)
    x_2 = x_1
    y_2 = y_1
    x_1 = x_0 - alpha*g_grad(x_0,y_0)[0]
    y_1 = y_0 - alpha*g_grad(x_0,y_0)[1]
    z_1 = g(x_1,y_1)
    ax.plot([x_0, x_1], [y_0, y_1], [z_0, z_1], color='black',zorder=10)
    x_0, y_0 = x_1, y_1
    if Cos(np.array([x_1-x_0,y_1-y_0]),np.array([x_2-x_1,y_2-y_1]))<-0.99:
        print('овраг',Cos(np.array([x_1-x_0,y_1-y_0]),np.array([x_2-x_1,y_2-y_1])))

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm, linewidth=0, antialiased=False)
ax = fig.add_subplot(1, 2, 2, projection='3d')

ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)
```
